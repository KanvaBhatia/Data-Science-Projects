## 1. Steps Game
- in a game of dice,
- if you toss a 1 or 2 you climb 1 step up the stairs
- if you toss 3,4,5 you climb 1 step down the stairs
- if you toss a 6, you toss again and climb the resulting number up the stairs
- There is a 0.1%  or 0.001 chance that you will fall; in which case, you will have to start again from bottom
- This program simulates the probability that you will reach the 60th step (after throwing 100 times)

Throw the dice, take a step forward or backward
- do this 100 time to see if you make it to the 60th step on the stairs
- to find the probability of making it to the 60 th step, repeat the 100 throw step a large number of times and count the number of times you make it to 60.
- account fo rthe possiilty of falling and erasing your cumulative count
- and if you throw a 3,4,5 when you are at the bottom just throw again. 

```python
# Set the seed
np.random.seed(123)
```

```python
# Empty list to contain the stair sequence for each 100 move run
stairs_climbed = []
```

```python
# repeat 100 throws 500 times
for i in range(500):
    
    # List to hold the results of each throw (in 100)  
    # will reset to 0 after each 100 throws (first step is at the bottom so 0)
    stair_per_throw = [0]
    
    # for 100 throws of the dice what step would you end on
    for x in range(100):
        current_stair = stair_per_throw[-1]
        
        # throw die
        die = np.random.randint(1,7)
        
        # set up control contruct: moving up/down stairs
        if die <= 2:
            current_stair += 1
        elif 3 <= die < 6:
            # throw a 3,4,5 you willl have to go a step down (current_stair - 1)
            # but if the current step is 0 that will be result in -1. hence use max to choose 0 
            current_stair = max(0, current_stair - 1)
        else:
            current_stair += np.random.randint(1,7)
            
        # account for possiblity of falling and reste step to 0 accordingly
        if np.random.rand() <= 0.001:
            current_stair = 0
            
        stair_per_throw.append(current_stair)
        
    stairs_climbed.append(stair_per_throw)
```

```python
np_stairs_climbed = np.array(stairs_climbed)
np_stairs_climbed
```

```python
plt.plot(np_stairs_climbed)
plt.show()
```

```python
# Transpose np_stairs_climbed
np_stairs_climbed_t = np.transpose(np_stairs_climbed)
np_stairs_climbed_t
```

```python
plt.plot(np_stairs_climbed_t)
plt.show()
```

- The last row (after transposing) represents the maximum no of stairs climbed at each of the 500, 100 throws

```python
# Select last row from np_stairs_climbed_t
last_row = np_stairs_climbed_t[100,:]
last_row
```

```python
plt.hist(last_row, bins = 20)
plt.show()
```

- To calculate the probabilty that one will reach 60 steps, we can count the no of integers int he last row that are greater than or equal to 60 and divide that by the total number of simultions (500)

```python
#The mean using only the integers that are >= 60
prob_60 = np.mean(last_row >= 60)
prob_60
```
