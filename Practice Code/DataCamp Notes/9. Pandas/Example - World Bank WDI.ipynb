- Generalise the code to select any row and any year

```python
def select_values(d_frame, row_ind, col_ind):
    columns = list(list(d_frame.columns[0:4]) + ['Year','Value'])
    
    select_row = wdi[row_ind:(row_ind + 1)].values.tolist()
    
    values = select_row[0][0:4] + [d_frame.columns[col_ind]] + [values_list[0][col_ind]]
    
    df = pd.DataFrame([values], columns = columns, index = [0])
    display(df)
```

- d_frame is wdi
- row index for adolescent fertility rate is 50 (row 51)
- year is 1960 so column index 4

```python
select_values(wdi, 50, 4)
```
