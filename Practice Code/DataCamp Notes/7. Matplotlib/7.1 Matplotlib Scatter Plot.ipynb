## 1. INTRODUCTION
- To use a scatter plot call plt.scatter(x, y). where x and y are coordinates
- Al customisations from Line Plots apply (labels, titles, color, marker etc. )

```python
# Creating dictionaries holding workers weekly hours

deshaun = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [1,2,3,4,5] }
aditya = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [1.5,2.3,3.5,4.2,7] }
mengfei = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [0.5,2.6,3.9,4.4,6] }
```

```python
# converting dictionaries to dataframes

deshaun_df = pd.DataFrame(deshaun)
aditya_df = pd.DataFrame(aditya)
mengfei_df = pd.DataFrame(mengfei)

display (deshaun_df)
```

```python
x = deshaun_df['weekday']
y = deshaun_df['hours']
x1 = aditya_df['weekday']
y1 = aditya_df['hours']
x2 = mengfei_df['weekday']
y2 = mengfei_df['hours']
```

- Scatter plot

```python
# LABELS AND TITLES
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

plt.scatter(x, y)
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```

## 2. Scatter Plot Customisations
Transparency, size

- #### Size (s)

```python
# LABELS AND TITLES
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

plt.scatter(x, y, s = 300)
# plt.scatter(x1, y1)
# plt.scatter(x2, y2)
plt.show()
```

- #### Transparency(alpha)
alpha goes fom 0 (transparent) to 1

```python
# LABELS AND TITLES
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

plt.scatter(x, y, s = 300, alpha = 0.1)
plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
```

## 3.Case Study
- explore the corellation life expectancy vs gdp per capita using a scatter plot

```python
# import dataset on countries

countries = pd.read_csv('Datasets/gapminder.csv')
countries.head()
```

- Create lists of the columns populations, life_exp and gdp_cap

```python
life_exp = countries['life_exp']
gdp_cap = countries['gdp_cap']
pop = countries.population
```

- Make a sctater plot to show relationship

```python
plt.scatter(gdp_cap, life_exp)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")
plt.show()
```

- Put x axis on a logaritmic scale 

```python
plt.scatter(gdp_cap, life_exp)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")

#log scale
plt.xscale('log')
plt.show()
```

- Create custom ticks for the x-axis

```python
plt.scatter(gdp_cap, life_exp)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")

#log scale
plt.xscale('log')

#ticks
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

plt.show()
```

- Represent the populations of each country using size on the scatter plot

```python
# the populaions are too large so we must divide across by 1,000,000. but firdt pop must become an np array

np_pop = np.array(pop)
np_pop /= 1000000
```

```python
# size
plt.scatter(gdp_cap, life_exp, s = np_pop)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")

#log scale
plt.xscale('log')

#ticks
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

plt.show()
```

- Apply Transparency

```python
# size, alpha
plt.scatter(gdp_cap, life_exp, s = np_pop, alpha = 0.5)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")

#log scale
plt.xscale('log')

#ticks
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

plt.show()
```

- Add color. 
- Since there are 147 countries adding each individual color would be tedious. hence we shall add color based on the continents which are 5 in total

```python
# Dictinct continents in the cont cont column

cont = countries.cont
cont.drop_duplicates()
```

```python
# create a dictionary to hold the colors

col = {'Asia': 'red',
      'Europe': 'green',
      'Africa': 'blue',
      'Americas': 'yellow',
       'Oceania': 'black'}
```

```python
# create a new list mapping all colors to continents
# add the dictionary(col) value for each continent in the list of continents 'cont'

colors = [ col[continent] for continent in cont]
```

```python
# size, alpha, color
plt.scatter(gdp_cap, life_exp, s = np_pop, alpha = 0.5, c = colors)

#labels
plt.xlabel("GDP Per Capita [in USD]")
plt.ylabel("Life Expectancy [in years]")
plt.title("World Development in 2007 [143 Countries]")

#log scale
plt.xscale('log')

#ticks
plt.xticks([1000, 10000, 100000], ['1k', '10k', '100k'])

plt.show()
```

- There is a general correlation between an increase in GDP and life expectancy
