## 1. Subquery Inside FROM

- Determine the number of languagses spoken in each country in the countries table

```python
%%sql
SELECT local_name, subquery.lang_count
FROM countries, 
    (SELECT code, COUNT(name) AS lang_count FROM languages
    GROUP BY code) AS subquery
WHERE countries.code = subquery.code
ORDER BY lang_count DESC
```

- Max percentage of women in parliament  for each continent in the states table

```python
%%sql
SELECT continent, MAX(women_parli_perc) AS max_perc
FROM states
GROUP BY continent
ORDER BY continent
```

- Max percentage of women in parliament  for each continent in the monarchs table

```python
%%sql
SELECT DISTINCT monarchs.continent, subquery.max_perc
FROM monarchs, 
    (SELECT continent, MAX(women_parli_perc) AS max_perc
    FROM states
    GROUP BY continent) AS subquery
WHERE monarchs.continent = subquery.continent
ORDER BY continent;
```
