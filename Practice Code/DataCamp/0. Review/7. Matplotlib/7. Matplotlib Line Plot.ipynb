## 2. Line Plots

```python
import pandas as pd
```

```python
# Creating dictionaries holding workers weekly hours

deshaun = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [8,5,3,5,8] }
aditya = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [10,2,1,0,0] }
mengfei = {'weekday': ['M', 'Tu', 'W', 'Th', 'F'], 'hours': [0,0,5,9,5] }
```

```python
# converting dictionaries to dataframes

deshaun_df = pd.DataFrame(deshaun)
aditya_df = pd.DataFrame(aditya)
mengfei_df = pd.DataFrame(mengfei)

display (deshaun_df)
```

```python
# defining axes

x = deshaun_df['weekday']
y = deshaun_df['hours']
x1 = aditya_df['weekday']
y1 = aditya_df['hours']
x2 = mengfei_df['weekday']
y2 = mengfei_df['hours']
```

```python
plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

## 3. Plot customisations

- #### Adding Axis Labels and title

```python
# LABELS AND TITLES
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

- #### Adding Legends
 -  add label argument to plt.plot(x,y,label ='')
 - you have to add plt.legend( ) just before plt.show() for it to work

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

# ADD LEGENDS
plt.plot(x, y, label = 'Deshaun')
plt.plot(x1, y1, label = 'Aditya')
plt.plot(x2, y2, label = 'Mengfei')

# ADD LEGEND
plt.legend()
plt.show()
```

- #### Adding text
plt.text(x-coord, y-coord, 'text here')

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

plt.plot(x, y, label = 'Deshaun')
plt.plot(x1, y1, label = 'Aditya')
plt.plot(x2, y2, label = 'Mengfei')

# ADD TEXT
plt.text(3, 2, 'Testing text')
plt.legend()
plt.show()
```

- #### Font Size
 - Add font size argument

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')

# FONT SIZE
plt.title('Time Sheet', fontsize = 20)

plt.plot(x, y, label = 'Deshaun')
plt.plot(x1, y1, label = 'Aditya')
plt.plot(x2, y2, label = 'Mengfei')

# FONT SIZE
plt.text(3, 2, 'Testing text', fontsize = 14)
plt.legend()
plt.show()
```

- #### Line Width

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

# CHANGE LINEWIDTH
plt.plot(x, y, label = 'Deshaun', linewidth = 4)
plt.plot(x1, y1, label = 'Aditya', linewidth = 4)
plt.plot(x2, y2, label = 'Mengfei', linewidth = 4)

plt.text(3, 2, 'Testing text')
plt.legend()
plt.show()
```

- #### Line Style

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

# LINE STYLE # Straight line linestyle = '-'
plt.plot(x, y, label = 'Deshaun', linestyle = ':')
plt.plot(x1, y1, label = 'Aditya', linestyle = '--')
plt.plot(x2, y2, label = 'Mengfei', linestyle = '-.')

plt.text(3, 2, 'Testing text')
plt.legend()
plt.show()
```

- #### Markers

```python
plt.xlabel('Weekday')
plt.ylabel('Hours Worked')
plt.title('Time Sheet')

# MARKERS # OTHERS d, *, h
plt.plot(x, y, label = 'Deshaun', marker = 'x')
plt.plot(x1, y1, label = 'Aditya', marker = 's')
plt.plot(x2, y2, label = 'Mengfei', marker = 'o')

# ADD TEXT
plt.text(3, 2, 'Testing text')
plt.legend()
plt.show()
```
